```python
casa['Vagas de Estacionamento'].fillna(0.0 , inplace= True)
casa['Área Construída'].fillna(casa['Área Construída'].mean() , inplace= True)
casa['Ano de Construção'].fillna(casa['Ano de Construção'].mean() , inplace= True)
```

```python
from sklearn.preprocessing import LabelEncoder
codificar = LabelEncoder() # da um nome para a funcao

casa["Bairro"] = codificar.fit_transform(casa["Bairro"])
casa["Tipo"] = codificar.fit_transform(casa["Tipo"])
casa["Método"] = codificar.fit_transform(casa["Método"])
casa["Vendedor"] = codificar.fit_transform(casa["Vendedor"])
casa["Região"] = codificar.fit_transform(casa["Região"])
casa.drop('Data' , axis=1 , inplace=True)
```

```python
#saparar a variavel alvo que e o valor q quer estimular, nesse caso o preco
# y como alvo 
# x como o resto

y = casa['Preço']
x = casa.drop('Preço' , axis= 1)
```

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
# criar o modelo 
modelo = ExtraTreesClassifier(n_estimators=100)
modelo.fit(x_treino , y_treino)

# imprimir os resultados

resultado = modelo.score(x_teste , y_teste)
print(resultado)
```
