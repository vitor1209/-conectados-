```python
import pandas as pd 
import numpy as np
```

```python
dados = pd.read_csv('onlinefraud.csv')
```

```python
display(dados)
```

```python
dados = dados.rename(columns={
    'step': 'etapa',
    'type': 'tipo',
    'amount': 'quantia',
    'nameOrig': 'nomeOrigem',
    'oldbalanceOrg': 'saldoAntigoOrigem',
    'newbalanceOrig': 'novoSaldoOrigem',
    'nameDest': 'nomeDestino',
    'oldbalanceDest': 'saldoAntigoDestino',
    'newbalanceDest': 'novoSaldoDestino',
    'isFraud': 'éFraude',
    'isFlaggedFraud': 'éFraudeMarcada'
})
```

```python
display(dados)
```

```python
dados.isnull().sum()
```

```python
dados.info()
```

```python
dados.tipo = dados.tipo.map({"CASH_OUT": 1, "PAYMENT": 2, "CASH_IN": 3, "TRANSFER": 4,"DEBIT": 5})

dados.head()
```

```python
# remove  == nomeOrigem , nomeDestino, éFraudeMarcada
removelist = dados[['nomeOrigem' , 'nomeDestino' , 'éFraude']]
dados = dados.drop(removelist , axis='columns')
```

```python
dados.head()
```

```python
dados.info()
```

```python
X = dados.drop('éFraudeMarcada' , axis='columns')
```

```python
y = dados['éFraudeMarcada']
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
clf = clf.fit(X_train, y_train)
```

```python
clf.score(X_train , y_train)
```

```python
y_pred = clf.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test , y_pred)
```

```python
from sklearn import metrics

metrics.accuracy_score(y_test , y_pred)
```

```python
features = np.array([[4, 9000.60, 9000.60, 0.0, 1.2, 3.4, 5.6]])
```

```python
res = clf.predict(features)
```
