```python
# NOTICIA 2
# Embalado por duas vitórias seguidas, o Corinthians entra em campo buscando ampliar a vantagem de pontos para a zona de rebaixamento e encerrar o primeiro turno de maneira positiva. Para isso, o Timão deve receber o apoio de mais de 40 mil pessoas e aproveitar o bom momento sob comando do técnico argentino Ramón Díaz.
```

```python
from sklearn.feature_extraction.text import CountVectorizer
doc = [str(news2)]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(doc)
```

```python
corpus = vectorizer.get_feature_names_out()
```

```python
corpus
```

```python
consulta = 'Qual é a atual situação do timão, Corinthians na tabela de classificação do campeonato após duas vitórias seguidas e qual a expectativa para o jogo com o apoio de mais de 40 mil torcedores do timão?'
```

```python
docConsu = [str(consulta)]
vectorizer2 = CountVectorizer()
C = vectorizer2.fit_transform(docConsu)
```

```python
consul = vectorizer2.get_feature_names_out()
consul
```

```python
from collections import Counter
cont = Counter(corpus)
cont
```

```python
contConsu = Counter(consul)
contConsu
```

```python
freq = []

for i in consul:
     for b in cont:
        if i == b:
            freq.append(i)
```

```python
cont = Counter(freq)
cont
```

```python
frequen = 0
for n in freq:
    frequen += 1
```

```python
contagem = 0
for n in corpus:
    contagem += 1
```

```python
import math
normaFreq = (frequen / contagem) * 100
score = math.log(normaFreq)
```
